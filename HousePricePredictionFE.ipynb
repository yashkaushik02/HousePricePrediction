# **Handling Missing Values**

```python
#checking the categoiral values
features_nan=[feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes=='O']

for feature in features_nan:
    print("{}: {}% missing values".format(feature,np.round(df[feature].isnull().mean(),4)))
```

```python
# Replace missing value with a new label

def replace_cat_feature(df, features_nan):
    data = df.copy()
    data[features_nan] = data[features_nan].fillna('Missing')
    return data

df = replace_cat_feature(df, features_nan)

print(df[features_nan].isnull().sum())
```

```python
#checking the numerical values
numerical_nan=[feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes!='O']

for feature in numerical_nan:
    print("{}: {}% missing values".format(feature,np.round(df[feature].isnull().mean(),4)))
```

```python
# Replace missing value with a new label

for feature in numerical_nan:
    median_value = df[feature].median()

    #creating a new fetaure to capture nan values
    df[feature+'nan'] = np.where(df[feature].isnull(), 1, 0)
    df[feature].fillna(median_value, inplace=True)

df[numerical_nan].isnull().sum()
```

```python
df.head(25)
```

```python
# Temporal Variables (Date time variables)

for feature in ['YearBuilt', 'YearRemodAdd','GarageYrBlt']:
    df[feature] = df['YrSold'] - df[feature]
```

# **Numerical Variables**
-> Since the numerical variables are skewed we will perform log noraml distribution

```python
df.head()
```

```python
#log normal distribution
num_features = ['LotFrontage','LotArea','1stFlrSF','GrLivArea','SalePrice']

for feature in num_features:
    df[feature] = np.log(df[feature])
```

```python
df.head()
```

# **Handling Rare Categorical Feature**
we will remove categorical variables that are present less than 1% of the observations

```python
categorical_features = [feature for feature in df.columns if df[feature].dtypes=='O']
categorical_features
```

```python
for feature in categorical_features:
    # Calculate the proportion of each category within the feature
    temp=df.groupby(feature)['SalePrice'].count()/len(df)

    # Select the categories that appear in more than 1% of the data
    temp_df=temp[temp>0.01].index

    # Replace categories that are not in the selected list with 'Rare_var'
    df[feature]=np.where(df[feature].isin(temp_df),df[feature],'Rare_var')
```

```python
df.head(100)
```

```python
for feature in categorical_features:
    labels_ordered=df.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    df[feature]=df[feature].map(labels_ordered)
```

# **Feature scaling**

```python
scaling_feature=[feature for feature in df.columns if feature not in ['Id','SalePerice'] ]
len(scaling_feature)
```

```python
scaling_feature
```

```python
#scaling the data between 0 to 1

from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(df[scaling_feature])
```

```python
scaler.transform(df[scaling_feature])
```

```python
#tranformed the train and test set and add on the ID and saleprice variables

data = pd.concat([df[['Id','SalePrice']].reset_index(drop = True),
                    pd.DataFrame(scaler.transform(df[scaling_feature]), columns = scaling_feature)],
                    axis=1)
```
